```python
# Title and Introduction
# ===================================
# Finding Heavy Traffic Indicators on I-94

# In this project, we analyze westbound traffic on the I-94 Interstate highway using exploratory data visualization techniques.
# The goal is to identify indicators that cause heavy traffic, such as time of day, day of the week, weather conditions, etc.
# We'll use pandas and matplotlib to visualize patterns and draw insights from the data.

# Step 1: Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Enable inline plots
%matplotlib inline

# Step 2: Load the Dataset
df = pd.read_csv('/content/Metro_Interstate_Traffic_Volume.csv')

# Step 3: View First and Last 5 Rows
print("First 5 Rows:")
print(df.head())
print("\nLast 5 Rows:")
print(df.tail())

# Step 4: Dataset Info
print("\nDataset Info:")
print(df.info())

# Step 5: Convert date_time to datetime format
df['date_time'] = pd.to_datetime(df['date_time'])

# Step 6: Add time-related columns
df['hour'] = df['date_time'].dt.hour
df['dayofweek'] = df['date_time'].dt.dayofweek
df['month'] = df['date_time'].dt.month
print("\n")

# Step 7: Visualize traffic volume by hour
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x='hour', y='traffic_volume', ci=None)
plt.title('Traffic Volume by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Traffic Volume')
plt.grid(True)
plt.show()
print("\n")

# Step 8: Traffic Volume by Day of Week
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='dayofweek', y='traffic_volume')
plt.title('Traffic Volume by Day of the Week')
plt.xlabel('Day of Week (0 = Monday)')
plt.ylabel('Traffic Volume')
plt.grid(True)
plt.show()
print("\n")

# Step 9: Traffic Volume by Weather
plt.figure(figsize=(14, 6))
sns.boxplot(data=df, x='weather_main', y='traffic_volume')
plt.title('Traffic Volume by Main Weather Type')
plt.xlabel('Weather Type')
plt.ylabel('Traffic Volume')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
print("\n")

# Step 10: Traffic Volume by Month
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x='month', y='traffic_volume')
plt.title('Traffic Volume by Month')
plt.xlabel('Month')
plt.ylabel('Traffic Volume')
plt.grid(True)
plt.show()
print("\n")

# Step 11: Traffic Volume Distribution on Holidays
plt.figure(figsize=(7, 5))
sns.boxplot(data=df, x='holiday', y='traffic_volume')
plt.title('Traffic Volume on Holidays vs Non-Holidays')
plt.xlabel('Holiday')
plt.ylabel('Traffic Volume')
plt.grid(True)
plt.show()
print("\n")

# Step 12: Correlation Heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(df[['traffic_volume', 'temp', 'rain_1h', 'snow_1h', 'clouds_all']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
